```python
# Rename the 'Name' column to 'Full Name'
df.rename(columns={'years_of_experience': 'experience'}, inplace=True)
```

Prepare the Data

```python
df[['salary', 'experience']].describe().round(2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
job_df = scaler.fit_transform(df[['salary', 'experience']])
```

```python
job_df = pd.DataFrame(job_df,
                      columns = ['salary', 'experience'])
```

```python
job_df.describe().round(2)
```

Create the Clusters

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters = 3, n_init = 15, random_state = 100)
```

```python
km.fit(job_df)
```

```python
km.labels_
```

```python
km.inertia_
```

 Evaluate and Visualize the Clusters

```python
pd.Series(km.labels_).value_counts().sort_index()
```

```python
km.cluster_centers_
```

```python
cluster_centers = pd.DataFrame(km.cluster_centers_,
                               columns = ['salary', 'experience'])
cluster_centers
```

```python
# Set the figure size
plt.figure(figsize=(10, 8))

#Plot the Clusters
ax = sns.scatterplot(data = job_df,
                     x = 'salary',
                     y = 'experience',
                     hue = km.labels_,
                     palette = 'colorblind',
                     alpha = 0.8,
                     s = 150,
                     legend = False)

#Plot the Centroids
ax = sns.scatterplot(data = cluster_centers,
                     x = 'salary',
                     y = 'experience',
                     hue = cluster_centers.index,
                     palette = 'colorblind',
                     s = 600,
                     marker = 'D',
                     ec = 'black',
                     legend = False)

# Add Centroid Labels
for i in range(len(cluster_centers)):
               plt.text(x = cluster_centers.salary[i],
                        y = cluster_centers.experience[i],
                        s = i,
                        horizontalalignment='center',
                        verticalalignment='center',
                        size = 15,
                        weight = 'bold',
                        color = 'white')
```

Choose the right number of Clusters

The Within Cluster Sum of Squares (WCSS)

```python
wcss = []
for k in range(2, 11):
    km = KMeans(n_clusters = k, n_init = 15, random_state = 100)
    km.fit(job_df)
    wcss.append(km.inertia_)

wcss_series = pd.Series(wcss, index = range(2, 11))

plt.figure(figsize=(8, 6))
ax = sns.lineplot(y = wcss_series, x = wcss_series.index)
ax = sns.scatterplot(y = wcss_series, x = wcss_series.index, s = 150)
ax = ax.set(xlabel = 'Number of Clusters (k)',
            ylabel = 'Within Cluster Sum of Squares (WCSS)')
```

The Average Silhouette Score

```python
from sklearn.metrics import silhouette_score

silhouette = []
for k in range(2, 11):
    km = KMeans(n_clusters = k, n_init = 15, random_state = 100)
    km.fit(job_df)
    silhouette.append(silhouette_score(job_df, km.labels_))

silhouette_series = pd.Series(silhouette, index = range(2, 11))

plt.figure(figsize=(8, 6))
ax = sns.lineplot(y = silhouette_series, x = silhouette_series.index)
ax = sns.scatterplot(y = silhouette_series, x = silhouette_series.index, s = 150)
ax = ax.set(xlabel = 'Number of Clusters (k)',
            ylabel = 'Average Silhouette Score')
```

The Calinski Harabasz Score

```python
from sklearn.metrics import calinski_harabasz_score

calinski = []
for k in range(2, 11):
    km = KMeans(n_clusters = k, n_init = 15, random_state = 100)
    km.fit(job_df)
    calinski.append(calinski_harabasz_score(job_df, km.labels_))

calinski_series = pd.Series(calinski, index = range(2, 11))

plt.figure(figsize=(8, 6))
ax = sns.lineplot(y = calinski_series, x = calinski_series.index)
ax = sns.scatterplot(y = calinski_series, x = calinski_series.index, s = 150)
ax = ax.set(xlabel = 'Number of Clusters (k)',
            ylabel = 'Calinski Harabasz Score')
```

Analyze and Interpret the Clusters

```python
km = KMeans(n_clusters = 5, n_init = 15, random_state = 100)
km.fit(job_df)
```

```python
pd.Series(km.labels_).value_counts().sort_index()
```

```python
cluster_centers = pd.DataFrame(km.cluster_centers_, columns = ['salary', 'experience'])
```

```python
# Set the figure size
plt.figure(figsize=(10, 8))

#Plot the Clusters
ax = sns.scatterplot(data = job_df,
                     x = 'salary',
                     y = 'experience',
                     hue = km.labels_,
                     palette = 'colorblind',
                     alpha = 0.8,
                     s = 150,
                     legend = False)

#Plot the Centroids
ax = sns.scatterplot(data = cluster_centers,
                     x = 'salary',
                     y = 'experience',
                     hue = cluster_centers.index,
                     palette = 'colorblind',
                     s = 600,
                     marker = 'D',
                     ec = 'black',
                     legend = False)


# Add Centroid Labels
for i in range(len(cluster_centers)):
               plt.text(x = cluster_centers.salary[i],
                        y = cluster_centers.experience[i],
                        s = i,
                        horizontalalignment='center',
                        verticalalignment='center',
                        size = 15,
                        weight = 'bold',
                        color = 'white')
```
