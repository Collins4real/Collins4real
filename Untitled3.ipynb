```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('Mobile_Price_Prediction_test.csv')
```

```python
df.head(10)
```

```python
df.describe()
```

```python
df.drop(['id'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
labels=df['four_g']
features=df.drop('four_g', axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)
```

```python
#  The parameter grid for GridSearchCV
param_grid = {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf'], 'gamma': [0.1, 0.5, 1.0]}
```

```python
# Creating GridSearchCV object with SVC model
svm_grid = GridSearchCV(SVC(), param_grid, cv=5, scoring='accuracy')
```

```python
# Fitting GridSearchCV to training data
svm_grid.fit(x_train, y_train)
```

```python
# Getting the best hyperparameters and best estimator
best_params = svm_grid.best_params_
best_estimator = svm_grid.best_estimator_
```

```python
# Making predictions on test data using the best estimator
y_pred = best_estimator.predict(x_test)
```

```python
# Evaluate model accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# Show best hyperparameters
print("Best Hyperparameters:", best_params)
```
